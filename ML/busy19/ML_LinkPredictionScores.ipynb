```python
#open the files and turn into list

import csv

with open('train.txt') as test:                                                                                          
    testz = csv.reader(test, delimiter='\t')
    data = list(testz)
    
with open('test-public.txt') as test2:                                                                                          
    test2z = csv.reader(test2, delimiter='\t')
    test_data = list(test2z)
```

```python
#import stuff
from sklearn.preprocessing import MinMaxScaler
import networkx as nx
import numpy as np
import pandas as pd
```

```python
import pickle
```

```python
with open('graph.pickle', 'rb') as f:
    G = pickle.load(f)
```

```python
#turn training data into a list of tuples; where each one represents a relationship
a = []
b = []
for i in range(len(data)):
    a.append(data[i][0])
#print(a[:2])

for i in range(2):
    print(len(data[i][1:]))
y = []
z = []
for i in range(len(data)):
    y = [a[i]]*len(data[i][1:])
    z.extend(list(zip(y,data[i][1:])))
```

```python
#look at what z looks like
z[0:5]
```

```python
len(z)
```

```python
import pickle

with open("./train.pickle", 'wb') as f:
    pickle.dump(z, f)
```

```python
#turn test data into a list of tuples like the list z above
source_test = []
for i in range(1,len(test_data)):
    source_test.append(test_data[i][1])

sink_test = []
for i in range(1,len(test_data)):
    sink_test.append(test_data[i][2])
```

```python
#check if all of test nodes is in training data
out = [item for t in z for item in t] 

out[0]

not_in = []

for i in source_test:
    if i not in out:
        not_in.append(i)
for i in sink_test:
    if i not in out:
        not_in.append(i)
        
len(not_in)
```

```python
#as above
test_tuple = list(zip(source_test,sink_test))
```

```python
#check test tuple
test_tuple
```

```python
#create a **undirected** graph G; because this stuff doesn't work with directed graphs..
G = nx.Graph()
```

```python
#add all the nodes, from our list z
G.add_edges_from(z)
```

```python
with open("./graph.pickle", 'wb') as f:
    pickle.dump(G, f)
```

```python
#calculate common neighbour score from test_tuple list
#Common neighbors captures the idea that two strangers who have a friend in common are more likely to be introduced than those who don’t have any friends in common.
#equation here https://neo4j.com/docs/graph-algorithms/current/algorithms/linkprediction-common-neighbors/
cn = [(e[0], e[1], len(list(nx.common_neighbors(G,e[0], e[1])))) for e in test_tuple]
```

```python
cn
```

```python
#computes Jaccard index
#The Jaccard index, also known as Intersection over Union and the Jaccard similarity coefficient (originally given the French name coefficient de communauté by Paul Jaccard), is a statistic used for gauging the similarity and diversity of sample sets. The Jaccard coefficient measures similarity between finite sample sets, and is defined as the size of the intersection divided by the size of the union of the sample sets
jc = list(nx.jaccard_coefficient(G, test_tuple))
```

```python
jc
```

```python
#computes resource allocation index
#Resource Allocation is a measure used to compute the closeness of nodes based on their shared neighbors.
ra = list(nx.resource_allocation_index(G,test_tuple))
```

```python
ra
```

```python
#computes adamic adar index
#similar to RA index above; but uses log
aa = list(nx.adamic_adar_index(G,test_tuple))
```

```python
aa
```

```python
#calculates preferred attachment - another measure of closeness but multiplies neighbours
pa = list(nx.preferential_attachment(G, test_tuple))
```

```python
pa
```

```python
#gen index for our test tuples
index = pd.Series([(e[0], e[1]) for e in cn])
```

```python
#initialise dataframe
df2 = pd.DataFrame()
```

```python
#extract score from the indexes calculated above
cn = pd.Series([(e[2]) for e in cn])
jc = pd.Series([(e[2]) for e in jc])
ra = pd.Series([(e[2]) for e in ra])
aa = pd.Series([(e[2]) for e in aa])
pa = pd.Series([(e[2]) for e in pa])
```

```python
#put scores in data frame
df2['common_neigh'] = cn
df2['jaccard'] = jc
df2['resource_alloc'] = ra
df2['adamic_adar'] = aa
df2['pref_attachment'] = pa
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#set index of dataframe as index of the test tuples
df2.index = index
```

```python
#This estimator scales and translates each feature individually such that it is in the given range on the training set, e.g. between zero and one.
scaler = MinMaxScaler()
```

```python
#scale the scores
scaled_df = pd.DataFrame(scaler.fit_transform(df2))
scaled_df.columns = ['common_neigh', 'jaccard', 'resource_alloc', 'adamic_adar', 'pref_attachment']
```

```python
#calculate mean of all scores
scaled_df['mean_score'] = (scaled_df['common_neigh'] + scaled_df['jaccard'] + scaled_df['resource_alloc']\
+ scaled_df['adamic_adar'] + scaled_df['pref_attachment'])/5
```

```python
#set scaled df index
scaled_df.index = index
```

```python
#looking at scaled table
scaled_df
```

```python
#looking at scaled table in descending order of scores
scaled_df.sort_values('jaccard', ascending=False)
```

```python
#the blocks below return different columns of the scaled table as scores; and outputs them as a csv
result = list(scaled_df['mean_score'])
```

```python
with open('wow.csv', 'w', newline='') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(result)
```
