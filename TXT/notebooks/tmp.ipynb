```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```python
dummy_fn = lambda x: x

tokens = [["hello", "there"], 
          ["cat", "hello", "dog", "mouse"]]

cv = CountVectorizer(tokenizer=dummy_fn,
                     preprocessor = dummy_fn,
                     ngram_range=(1,1),
                     max_features=4)

tfidf = TfidfTransformer()

cv.fit(tokens)
C = cv.transform(tokens)
print("Count matrix:")
print(cv.get_feature_names())
print(C.todense())

tfidf.fit(C)
X = tfidf.transform(C)

print("\ntf-idf matrix:")
print(X.todense())


# tranform some new text
new_text = ["hello", "cat", "ball", "mouse", "mouse", "hello", "tissues"]

N = cv.transform([new_text]).todense()
idfN = tfidf.transform(N)

print("\nNew text counts matrix:")
print(N)

print("\nNew text tf-idf matrix:")
print(idfN.todense())
```
