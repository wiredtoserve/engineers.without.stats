```python
def merge_col(row):
    post = []
    if type(row[4]) == float:
        post = str(row[2]) + ' ' + str(row[3]) + ' ' + str(row[5])
    else:
        post = str(row[2]) + ' ' + str(row[3]) + ' ' + str(row[4])
        
    return post
```

## MBTI Scoring

```python
response = pd.read_excel('MBTI questions (Responses).xlsx')
```

```python
response.head()
```

```python
from collections import defaultdict
def count_type_1(row):
    my_dd = defaultdict(int)
    iterate_list = [1,8,15,22,29,36,43,50,57,64]
    for i in iterate_list:
        my_dd[row[i+1].split(')')[0]] += 1
    return my_dd
```

```python
response['Type_1'] = response.apply(count_type_1, axis=1)
```

```python
def count_type_2(row):
    my_dd = defaultdict(int)
    iterate_list = [2,9,16,23,30,37,44,51,58,65,3,10,17,24,31,38,45,52,59,66]
    for i in iterate_list:
        my_dd[row[i+1].split(')')[0]] += 1
    return my_dd
```

```python
response['Type_2'] = response.apply(count_type_2, axis=1)
```

```python
def count_type_3(row):
    my_dd = defaultdict(int)
    iterate_list = [4,11,18,25,32,39,46,53,60,67,5,12,19,26,33,40,47,54,61,68]
    for i in iterate_list:
        my_dd[row[i+1].split(')')[0]] += 1
    return my_dd
```

```python
response['Type_3'] = response.apply(count_type_3, axis=1)
```

```python
def count_type_4(row):
    my_dd = defaultdict(int)
    iterate_list = [6,13,20,27,34,41,48,55,62,69,7,14,21,28,35,42,49,56,63,70]
    for i in iterate_list:
        my_dd[row[i+1].split(')')[0]] += 1
    return my_dd
```

```python
response['Type_4'] = response.apply(count_type_4, axis=1)
```

```python
cols = [x for x in response.columns]
```

```python
score_df = response.drop(cols, axis=1)
```

```python
score_df
```

```python
score_df['E/I'] = score_df['Type_1'].apply(lambda x:'E' if x['a'] > x['b'] else 'I')
```

```python
score_df['S/N'] = score_df['Type_2'].apply(lambda x:'S' if x['a'] > x['b'] else 'N')
```

```python
score_df['T/F'] = score_df['Type_3'].apply(lambda x:'T' if x['a'] > x['b'] else 'F')
```

```python
score_df['J/P'] = score_df['Type_4'].apply(lambda x:'J' if x['a'] > x['b'] else 'P')
```

```python
score_df
```

```python
final_dataset = pd.merge(text_response, score_df, left_on="Full Name", right_on="Full name", how="inner")
```
