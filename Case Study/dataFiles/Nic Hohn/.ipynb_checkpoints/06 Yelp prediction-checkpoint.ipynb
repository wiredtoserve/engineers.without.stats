```
# create a DataFrame from the 'reviews' table and print its schema
reviews = spark.sql("select * from reviews").cache()
reviews.printSchema()
```

```
import string
import re


from pyspark.sql.functions import udf
from pyspark.ml.feature import *
from pyspark.mllib.classification import SVMWithSGD
from pyspark.mllib.regression import LabeledPoint
from pyspark.mllib.linalg import Vector as MLLibVector, Vectors as MLLibVectors
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import BinaryClassificationMetrics
from pyspark.sql.types import IntegerType

import pandas as pd

n_reviews = 1000000
min_pct = 0.0001
min_count = n_reviews * min_pct

# remove punctuation
def remove_punct(text):
    regex = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n]')
    nopunct = regex.sub(" ", text)
    return nopunct

# binarize rating
def convert_rating(rating):
    rating = int(rating)
    if rating >=4: return 1
    else: return 0

# udf
punct_remover = udf(lambda x: remove_punct(x))
rating_convert = udf(lambda x: convert_rating(x))

# apply to review raw data
review_df = reviews.select('review_id', punct_remover('text'), rating_convert('stars'))

review_df = review_df.withColumnRenamed('<lambda>(text)', 'text')\
                     .withColumn('label', review_df["<lambda>(stars)"].cast(IntegerType()))\
                     .drop('<lambda>(stars)')\
                     .limit(n_reviews)

review_df.show(5)
```

```
# tokenize
tok = Tokenizer(inputCol="text", outputCol="words")
review_tokenized = tok.transform(review_df)

# remove stop words
stopword_rm = StopWordsRemover(inputCol='words', outputCol='words_nsw')
review_tokenized = stopword_rm.transform(review_tokenized)

review_tokenized.show(5)
```

```
# add ngram column
n = 3
ngram = NGram(inputCol = 'words_nsw', outputCol = 'ngram', n = n)
add_ngram = ngram.transform(review_tokenized)

add_ngram.show(5)
```

```
# generate the top frequent ngram
ngrams = add_ngram.rdd.flatMap(lambda x: x[-1]).filter(lambda x: len(x.split())==n)
ngram_tally = ngrams.map(lambda x: (x, 1))\
                      .reduceByKey(lambda x,y: x+y)\
                      .sortBy(lambda x: x[1], ascending=False)\
                      .filter(lambda x: x[1] >= min_count)
ngram_list = ngram_tally.map(lambda x: x[0]).collect()
```

```
# replace the word with selected ngram
def ngram_concat(text):
    text1 = text.lower()
    for ngram in ngram_list:
        return text1.replace(ngram, ngram.replace(' ', '_'))

ngram_df = udf(lambda x: ngram_concat(x))
ngram_df = review_tokenized.select(ngram_df('text'), 'label')\
                          .withColumnRenamed('<lambda>(text)', 'text')

# tokenize and remove stop words with ngram
tok = Tokenizer(inputCol="text", outputCol="words")
review_tokenized = tok.transform(review_df)
tokenized_ngram = tok.transform(ngram_df)
tokenized_ngram = stopword_rm.transform(tokenized_ngram)

stopword_rm = StopWordsRemover(inputCol='words', outputCol='words_nsw')
review_tokenized = stopword_rm.transform(review_tokenized)

# count vectorizer and tfidf
cv = CountVectorizer(inputCol='words_nsw', outputCol='tf')
cvModel = cv.fit(review_tokenized)
count_vectorized = cvModel.transform(review_tokenized)

idf_ngram = IDF().setInputCol('tf').setOutputCol('tfidf')
tfidfModel = idf_ngram.fit(count_vectorized)
tfidf_df = tfidfModel.transform(count_vectorized)
```

```
# split into training and testing set (always fix the seed!)
splits = tfidf_df.select(['tfidf', 'label']).randomSplit([0.8,0.2], seed=100)
train = splits[0].cache()
test = splits[1].cache()

# Convert feature matrix to LabeledPoint vectors
train_lb = train.rdd.map(lambda row: LabeledPoint(row[1], MLLibVectors.fromML(row[0])))
test_lb = test.rdd.map(lambda row: LabeledPoint(row[1], MLLibVectors.fromML(row[0])))

# SVM model
numIterations = 50
regParam = 0.3
svm = SVMWithSGD.train(train_lb, numIterations, regParam=regParam)

scoreAndLabels_test = test_lb.map(lambda x: (float(svm.predict(x.features)), x.label))
score_label_test = spark.createDataFrame(scoreAndLabels_test, ["prediction", "label"])

# Instantiate metrics object
metrics = BinaryClassificationMetrics(scoreAndLabels_test)

# Area under ROC curve
print("Area under ROC = %s" % metrics.areaUnderROC)

# F1 score
f1_eval = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="f1")
svm_f1 = f1_eval.evaluate(score_label_test)
print("F1 score: %.4f" % svm_f1)

vocabulary_ngram = cvModel.vocabulary
weights_ngram = svm.weights.toArray()
svm_coeffs_df_ngram = pd.DataFrame({'ngram': vocabulary_ngram, 'weight': weights_ngram})

words = dict()
words['negative'] = svm_coeffs_df_ngram.sort_values('weight').head(40).set_index('ngram').apply(lambda x: -x).to_dict()['weight']
words['positive'] = svm_coeffs_df_ngram.sort_values('weight', ascending=False).head(40).set_index('ngram').to_dict()['weight']
```

```
import matplotlib.pyplot as plt
from wordcloud import WordCloud
fig = {}
i = 0
for key in words.keys():
    wordcloud = WordCloud(max_font_size=150, max_words=200, background_color="white", width=900, height=600) \
                        .generate_from_frequencies(words[key])
    # Display the generated image:
    fig[i], ax = plt.subplots()
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis("off")
    plt.show()
    i += 1
```

```
# import data to pandas dataframes
import pandas as pd

X = spark.sql("select categories, review_count, attributes.* from business").toPandas()
y = spark.sql("select stars from business").toPandas()
y = y['stars'].apply(lambda x: int(x >= 4))

col_cat = X.select_dtypes(include=['object', 'category']).columns.tolist()

# Recode
for col in col_cat:
    # Convert column to pd.Categorical
    X[col] = pd.Categorical(X[col])
    # Extract the cat.codes and replace the column with these
    X[col] = X[col].cat.codes
    # Convert the cat codes to categorical...
    X[col] = pd.Categorical(X[col])
```

```
# setup a lightGBM predictive model
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import (roc_curve, roc_auc_score, accuracy_score)
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

d_train = lgb.Dataset(X_train, label = y_train, free_raw_data=False)
d_test = lgb.Dataset(X_test, label = y_test, free_raw_data=False)
```

```
params = {'boosting_type': 'gbdt',
          'max_depth' : -1,
          'objective': 'binary',
          'nthread': 3, # Updated from nthread
          'num_leaves': 64,
          'learning_rate': 0.05,
          'max_bin': 512,
          'subsample_for_bin': 200,
          'subsample': 0.75,
          'subsample_freq': 1,
          'colsample_bytree': 0.8,
          'reg_alpha': 1.2,
          'reg_lambda': 1.2,
          'min_split_gain': 0.5,
          'min_child_weight': 1,
          'min_child_samples': 5,
          'scale_pos_weight': 1,
          'num_class' : 1,
          'metric' : 'auc',
           }
```

```
import warnings
with warnings.catch_warnings():
    model = lgb.train(params, d_train, categorical_feature=col_cat,valid_sets=d_train,verbose_eval=False)    
    trainPerformance = roc_auc_score(y_train, model.predict(X_train))
    testPerformance = roc_auc_score(y_test, model.predict(X_test))

print("Model train performance: AUC=%.02f" % trainPerformance)
print("Model test performance:  AUC=%.02f" % testPerformance)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

feature_imp = pd.DataFrame(sorted(zip(model.feature_importance(),X_train.columns)), columns=['Value','Feature'])

plt.figure(figsize=(20, 10))
sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
plt.title('LightGBM Features (avg over folds)')
plt.tight_layout()
display()
```
