```python
def count_items(x):
    from collections import defaultdict
    mydd = defaultdict(int)
    
    for i in x:
        mydd[i] += 1
        
    return mydd
```

```python
state_list = []
for key, val in state_dict.items():
    if val > 10:
        state_list.append(key)
```
